```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
%matplotlib inline
```

```python
url = 'https://github.com/books-by-chansung/low-code-ai/blob/main/advertising_2023.csv?raw=true'
advertising_df = pd.read_csv(url,index_col=0)
advertising_df.head()
```

```python
advertising_df.info()
```

```python
advertising_df.describe()
```

```python
advertising_df.isnull().sum()
```

```python
advertising_df.isnull().values.any()
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(advertising_df.corr(), annot=True, vmin=0, vmax=1, cmap='viridis')
```

```python
advertising_df.plot(kind='scatter', x=['digital'], y='sales')
```

```python
plt.figure(figsize=(18, 18))

for i, col in enumerate(advertising_df.columns[0:13]):
  plt.subplot(5, 3, i+1)
  x = advertising_df[col]
  y = advertising_df['sales']
  plt.plot(x, y, 'o')

  plt.plot(np.unique(x), np.poly1d(np.polyfit(x, y, 1)) (np.unique(x)), color='red')
  plt.xlabel(col)
  plt.ylabel('sales')
```

```python
sns.displot(advertising_df, x="sales")
```

```python
lis = ['digital', 'newspaper', 'radio', 'TV']
plt.subplots(figsize=(15, 8))
index = 1
for i in lis:
  plt.subplot(2, 2, index)
  sns.distplot(advertising_df[i])
  index += 1
```
